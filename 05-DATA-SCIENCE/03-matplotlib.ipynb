âœ… Basic Line Plot

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 14, 7]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
````

âœ… Bar Chart

````python
import matplotlib.pyplot as plt

names = ["Alice", "Enzo", "Charlie"]
scores = [85, 90, 78]

plt.bar(names, scores, color="purple", edgecolor = "red")
plt.title("Scores by Person")
plt.ylabel("Score")
plt.show()
````

âœ… Scatter Plot

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color="green")
plt.title("Random Scatter Plot")
plt.show()
````

âœ… Histogram

````python
import matplotlib.pyplot as plt

data = [22, 25, 20, 21, 30, 35, 29, 30, 30, 25, 28]

plt.hist(data, bins=5, color="skyblue", edgecolor="black")
plt.title("Histogram of Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
````

âœ… Plot from Pandas DataFrame

````python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Name": ["Anna", "Brian", "Cecilia", "Daniel"],
    "Score": [88, 92, 79, 85]
}
df = pd.DataFrame(data)

plt.bar(df["Name"], df["Score"], color="purple")
plt.title("Scores by Student")
plt.ylabel("Score")
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Months": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [1500, 1800, 1700, 1600, 2100, 1900]
}

df = pd.DataFrame(data)

plt.plot(df["Months"], df["Sales"], color = "skyblue", marker="o")
plt.title("Sales over time")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.grid(True)
plt.show()

plt.bar(df["Months"], df["Sales"], color = "skyblue")
plt.title("Sales per month")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True, axis='y')
plt.show()
````
